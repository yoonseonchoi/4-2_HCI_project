```python
import pandas as pd
import numpy as np
```

```python
makeUp_df = pd.read_csv('makeUp.csv')
makeUp_df['피부타입'] = makeUp_df['피부타입'].astype('string')

temp = []
for i in range(len(makeUp_df['피부타입'])):
    temp.append(eval(makeUp_df['피부타입'][i]))    
makeUp_df['피부타입'] = temp

makeUp_df
```

```python
for i in makeUp_df['피부타입']:
    i['지성'] = int(round(i['지성'] / (i['지성'] + i['건성'] + i['복합성']) * 100, 0))
    i['건성'] = int(round(i['건성'] / (i['지성'] + i['건성'] + i['복합성']) * 100, 0))
    i['복합성'] = int(round(i['복합성'] / (i['지성'] + i['건성'] + i['복합성']) * 100, 0))
```

```python
makeUp_df.to_csv('./makeUp.csv', index = False, encoding='utf-8-sig')
```
