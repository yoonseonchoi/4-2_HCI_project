## User Information

```python
file = open(r'./user_info.txt')
user_info=file.readlines()

name=user_info[0]
name=name.strip()
print("--- "+str(name)+"님의 정보 ---")

age=user_info[1]
age = age.strip()
print('age: ',age)

skinType=user_info[2]
skinType=skinType.strip()
print('skinType: ',skinType)

skinConcern=user_info[3]
print('skinConcern: ',skinConcern)
```

## Import libraries

```python
import pandas as pd
import re
```

## Dataset Upload

```python
Cleansing=pd.read_csv('./cleansing.csv',encoding='cp949')
pd.set_option('display.max_rows',None)
```

```python
Ranked = Cleansing[Cleansing['나이'].str.contains(age)]
Ranked = Ranked[Ranked['피부타입'].str.contains(skinType)]

#If considered 2 above, but if length is 0: consider only skinType
if len(Ranked)==0:
    Ranked =Cleansing[Cleansing['피부타입'].str.contains(skinType)]
    
# others
others=pd.merge(Ranked, Cleansing, how='outer',indicator=True)
others=others.query('_merge=="right_only"').drop(columns=['_merge'])

#reset index
Ranked.reset_index(inplace=True)
Ranked = Ranked.drop(['index', '랭킹','리뷰수'], axis=1)
Ranked.index=Ranked.index + 1
others.reset_index(inplace=True)
others = others.drop(['index', '랭킹','리뷰수'], axis=1)
others.index=others.index + 1
```

## OCR

```python
ocr_result=[]

with open("./ocr.txt", "r") as ocr_file:
    for line in ocr_file:
        line=line.strip()
        ocr_result.append(line)

print("OCR Result: ",ocr_result)
```

## Recommendation Ranking after OCR

```python
OCR_Ranked = Ranked
OCR_Ranked['주의성분유무'] = 'X'
for i in range (len(ocr_result)):
    OCR_Ranked_index=OCR_Ranked[OCR_Ranked['성분'].str.contains(ocr_result[i])==True]
OCR_Ranked_index1=OCR_Ranked_index.reset_index(drop=False)
```

```python
indexes=OCR_Ranked_index1['index']
index_list=indexes.values
```

```python
for i in range (len(index_list)):
    OCR_Ranked.loc[index_list[i], '주의성분유무'] = '주의'
OCR_Ranked.reset_index(inplace=True)
OCR_Ranked.index=OCR_Ranked.index + 1
print("주의상품개수:",len(OCR_Ranked_index))
```

```python
OCR_Ranked = OCR_Ranked.drop(['index'], axis=1)
OCR_Ranked = OCR_Ranked[['상품명','주의성분유무','가격','성분','나이','피부타입']]
OCR_Ranked
```
